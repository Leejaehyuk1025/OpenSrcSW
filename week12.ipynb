```python
#실습1. 서울의 일교차
import csv
f1 = open('seoul.csv')
data1= csv.reader(f1)
next(data1)

f2 = open('seoul.csv')
data2= csv.reader(f2)
next(data2)

highTemp = float(0) #최고기온이 가장 높은날
lowTemp = float(999) #최저기온이 가장  낮은날
bigGapTemp = float(0) #일교차가 가장 큰 날

for row in data1:
    row[3] = float(row[3])
    row[4] = float(row[4])
    if lowTemp > row[3]:
         lowTemp = row[3]
    if row[4] > highTemp:
         highTemp = row[4]
    temp = float(row[4]) - float(row[3])
    if bigGapTemp < temp:
         bigGapTemp = temp
    
highDate = ''
lowDate = ''
bigGapDate = ''

for row in data2:
    row[3] = float(row[3])
    row[4] = float(row[4])
    if float(row[4]) == float(highTemp):
        highDate = row[0]
    if row[3] == lowTemp:
        lowDate = row[0]
    
    temp = row[4] - row[3]
    if temp == bigGapTemp:
        bigGapDate = row[0]
    
print("일교차가 가장 큰 날은 ", bigGapDate, "입니다.")
print("최저기온이 가장 낮은 날은 ",lowDate," 입니다.")
print("최고기온이 가장 높은 날은 ",highDate," 입니다.")
f1.close()
f2.close()
```

```python
#실습2. 지하철 이용객
import csv
f = open('subwayfee.csv')
data= csv.reader(f)
next(data)

listName = ['1호선','2호선','3호선','4호선','5호선','6호선','7호선','8호선','9호선','9호선2~3단계','경부선','경인선','경원선',
           '안산선','과천선','분당선','일산선','중앙선','장항선','경의선','경춘선','수인선','경강선','공항철도 1호선','우이신설선']

for station in listName:
    dictionary = {}
    f = open('subwayfee.csv')
    data= csv.reader(f)
    next(data)
    for row in data:
        if row[1] == station:
            temp = int(row[4].replace(',','')) + int(row[5].replace(',','')) + int(row[6].replace(',','')) + int(row[7].replace(',',''))
            dictionary[temp] = row[3]
        
    list_a = list(dictionary.keys())
    list_a.sort()
    #print(list_a)
    print("============================================")
    print(station, ": 이용객이 가장 많은 역 순위")
    print("       1. ",dictionary[list_a[len(list_a)-1]],list_a[len(list_a)-1])
    print("       1. ",dictionary[list_a[len(list_a)-2]],list_a[len(list_a)-2])
    print("       1. ",dictionary[list_a[len(list_a)-3]],list_a[len(list_a)-3])
    print(station, ": 이용객이 가장 적은 역 순위")
    print("       1. ",dictionary[list_a[0]],list_a[0])
    print("       1. ",dictionary[list_a[1]],list_a[1])
    print("       1. ",dictionary[list_a[2]],list_a[2])


f.close()
```
