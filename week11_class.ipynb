```python
#실습 8. 사칙 연산 클래스
class Calc:
    def sum(self,a,b):
        result = a + b
        print("{0:.3f} + {1:.3f} = {2:.3f} 입니다".format(a,b,result))
    def sub(self,a,b):
        result = a - b
        print("{0:.3f} - {1:.3f} = {2:.3f} 입니다".format(a,b,result))
    def mul(self,a,b):
        result = a * b
        print("{0:.3f} * {1:.3f} = {2:.3f} 입니다".format(a,b,result))
    def div(self,a,b):
        result = a / b
        print("{0:.3f} / {1:.3f} = {2:.3f} 입니다".format(a,b,result))
```
