```python
#문제1

class Point:
    x = 0
    y = 0
    
    
    def __add__(self,other):
        self.x = self.x+other.x
        self.y = self.y+other.y
        return Point(self.x,self.y)
    
    def __init__(self,inputX,inputY):
        self.x = inputX
        self.y = inputY
        
    def show(self):
        print((self.x , self.y))
        
    
    
    
a = Point(1,2)
a.show()

b=Point(3,4)
b.show()

a=a+b
a.show()
```
