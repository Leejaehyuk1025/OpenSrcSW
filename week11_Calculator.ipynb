```python
userinput = input("계산식 입력: ")
expression = userinput.split()

number = stack()
operator = stack()


for i in range(0,len(expression),2):
    if expression[i-1] == '*':
        x = number.pop()
        operator.pop()
        number.push(int(expression[i]) * int(x))
    else:
        number.push(expression[i])
    
    if i + 1 == len(expression):
        break
    else:
        i += 1
        operator.push(expression[i])
    

result = 0

for j in range(0,operator.len(),1):
    temp1 = operator.pop()
    temp2 = int(number.pop())
    if temp1 == '+':
        result = result + temp2
    elif temp1 == '-':
        temp2 = 0 - temp2
        result = result + temp2
    else:
        print("error")
        

result = result + int(number.pop())
for i in range(0,len(expression),1):
    print(expression[i], end = ' ')
print(" 의 연산 결과는 ",result, " 입니다.")
```
