```python
#최대값 구하기

x = input("Numbers? : ")
a=x.split()
max = int(a[0])
for i in a:
    if int(i) > max:
        max=int(i)
print('MAX :',max)
```

```python
#암호화/복호화

x = input("1:암호화 2:복호화 >>")

if x == "1":
    y = input("암호화 할 문자열을 입력하세요")
    ylist = y.split()
    ylist.reverse()
    result=""
    for i in ylist:
        newstring = ""
        for j in i:
            code = ord(j)
            j = chr(code + 1)
            newstring += j
        result += newstring + " "
    print(result)
    
elif x == "2":
    z = input("복호화 할 문자열을 입력하세요")
    zlist = z.split()
    newlist=[]
    result=""
    for i in zlist:
        newstring = ""
        for j in i:
            code = ord(j)
            j = chr(code - 1)
            newstring += j
        newlist.append(newstring)
    newlist.reverse()
    for k in newlist:
        result += k + " "
    print(result)
    
else:
    print('잘못된 입력입니다.')
```

```python
#전화번호부

dic={}
newdic={}

for i in range(5):
    a=input("이름과 전화번호를 입력하세요")
    alist = a.split()
    dic[alist[0]]=alist[1]

select = input("1:이름 검색 2:전화번호 검색")

if select == "1":
    name = input("이름을 입력하세요: ")
    if name in dic:
        print("전화번호는 ",dic.get(name)," 입니다.")
    else:
        print("존재하지 않는 이름입니다.")
        
elif select == "2":
    phone = input("전화번호를 입력하세요: ")
    newdic = {v:k for k,v in dic.items()}
    if phone in newdic:
        print("이름은 ",newdic.get(phone)," 입니다.")
    else:
        print("존재하지 않는 전화번호입니다.")
        
else:
    print("잘못된 입력입니다.")
```
