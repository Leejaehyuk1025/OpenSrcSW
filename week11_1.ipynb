```python
#실습6. Copy

userinput = input()
a = userinput.split()
if a[0] == 'cp':
    file1 = open(a[1],'r')
    file2 = open(a[2],'w')
    
    data = file1.read()
    file2.write(data)
    
    file1.close()
    file2.close()
    print('파일 복사 완료.')
else:
    print('잘못된 입력입니다.')
```

```python
#실습7. Wc

userinput = input()
a = userinput.split()
if a[0] == 'wc':
    file1 = open(a[1],'r')
    
    countWords = 0
    countLines = 0
    
    for line in file1:
        line = line.replace('\n','')
        words = line.split()
        for word in words:
            countWords += 1
        countLines += 1    
    
    print("파일의 단어 수는 {0:d} 입니다.".format(countWords))
    print("파일의 줄 수는 {0:d} 입니다.".format(countLines))
    
    file1.close()
    
else:
    print('잘못된 입력입니다.')
```
