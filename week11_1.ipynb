```python
#실습6. Copy

userinput = input()
a = userinput.split()
if a[0] == 'cp':
    file1 = open(a[1],'r')
    file2 = open(a[2],'w')
    
    data = file1.read()
    file2.write(data)
    
    file1.close()
    file2.close()
    print('파일 복사 완료.')
else:
    print('잘못된 입력입니다.')
```

```python
#실습7. Wc

userinput = input()
a = userinput.split()
if a[0] == 'wc':
    file1 = open(a[1],'r')
    
    countWords = 0
    countLines = 0
    
    for line in file1:
        line = line.replace('\n','')
        words = line.split()
        for word in words:
            countWords += 1
        countLines += 1    
    
    print("파일의 단어 수는 {0:d} 입니다.".format(countWords))
    print("파일의 줄 수는 {0:d} 입니다.".format(countLines))
    
    file1.close()
    
else:
    print('잘못된 입력입니다.')
```

```python
#실습9. 계산기

userinput = input("계산식 입력: ")
a = userinput.split()

numList = list()
calList = list()

order = 0
for x in a:
    
    if order == 1:
        num1 = int(x)
        num2 = int(numList.pop())
        numList.append(num1*num2)
        order = 0
    elif order == 2:
        num1 = int(x)
        num2 = int(numList.pop())
        numList.append(num1/num2)
        order = 0
    else:
        if x == '+' :
            calList.append(x)
        elif x == '-' :
            calList.append(x)
        elif x == '*' :
            calList.append(x)
            order = 1
        elif x == '/' :
            calList.append(x)
            order = 2
        else:
            numList.append(x)
            
            
while len(calList):
    num1 = int(numList.pop())
    num2 = int(numList.pop())
    num3 = calList.pop()
    
    if(num3 == '+'):
        numList.append(num1+num2)
    if(num3 == '-'):
        numList.append(num1+num2)

print("{0:f}".format(numList[0]))
```
