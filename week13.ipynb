```python
import matplotlib.pyplot as plt
import csv
f = open('seoul1.csv')
data = csv.reader(f)
next(data)

highTemp = []
lowTemp = []
for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '25':
            highTemp.append(float(row[4]))
            lowTemp.append(float(row[3]))
            
plt.plot(highTemp, color = 'red', label = 'Highest')
plt.plot(lowTemp, color = 'blue', label = 'Lowest')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul1.csv')
data = csv.reader(f)
next(data)
january = []
august = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '01':
            january.append(float(row[4]))
        if month == '08':
            august.append(float(row[4]))
            
plt.hist(august, bins=100, color='r', label='Aug')            
plt.hist(january, bins=100, color='b', label='Jan')

plt.legend()
plt.show()
```
